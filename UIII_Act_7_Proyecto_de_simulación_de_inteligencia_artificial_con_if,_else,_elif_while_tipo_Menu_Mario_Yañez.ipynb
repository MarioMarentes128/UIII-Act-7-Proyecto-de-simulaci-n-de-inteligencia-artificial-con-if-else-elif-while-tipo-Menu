```python
while True:
    print("\n--- JOYERÍA INTELIGENTE ---")
    print("1. Recomendar joya")
    print("2. Ver precios")
    print("3. Elegir regalo")
    print("4. Salir")

    opcion = input("Selecciona una opción: ")

    if opcion == "1":
        tipo = input("¿Te gusta oro, plata o diamante? ")

        if tipo == "oro":
            print("Te recomendamos un anillo de oro.")
        elif tipo == "plata":
            print("Te recomendamos una cadena de plata.")
        elif tipo == "diamante":
            print("Te recomendamos un collar con diamante.")
        else:
            print("No tenemos ese tipo de joya.")

    elif opcion == "2":
        print("Anillo de oro: $5000")
        print("Cadena de plata: $2500")
        print("Collar diamante: $12000")

    elif opcion == "3":
        regalo = input("¿Es para hombre o mujer? ")

        if regalo == "hombre":
            print("Recomendación: reloj elegante.")
        elif regalo == "mujer":
            print("Recomendación: collar brillante.")
        else:
            print("Opción no válida.")

    elif opcion == "4":
        print("Gracias por visitar la joyería.")
        break

    else:
        print("Opción incorrecta.")
print("Programa realizado por Mario Yañez NL = 53")
```
